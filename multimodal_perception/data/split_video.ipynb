```python
import pandas as pd
import subprocess
from pathlib import Path
```

```python
# -----------------------
# Paths
# -----------------------

csv_path = "pilot.csv"                  # original csv
audio_root = Path("../../../pilot/")    # folder containing date folders
output_dir = Path("../../assets/video/pilot_v2")

output_dir.mkdir(parents=True, exist_ok=True)

# -----------------------
# Load CSV
# -----------------------

df = pd.read_csv(
    csv_path,
    dtype={
        "Clue audio start": str,
        "Clue audio end ": str
    }
)

# add clue ids
df["clue_id"] = range(1, len(df) + 1)

# convert date format (27/02/2026 -> 27-02-2026)
df["date_folder"] = pd.to_datetime(df["Date "], dayfirst=True).dt.strftime("%d-%m-%Y")


# -----------------------
# Time parser
# -----------------------

def to_seconds_ms(t):
    if pd.isna(t):
        return None
    t = str(t).strip()
    try:
        parts = list(map(float, t.split(":")))
        if len(parts) == 2:  # mm:ss
            return parts[0]*60 + parts[1]
        elif len(parts) == 3:  # hh:mm:ss
            return parts[0]*3600 + parts[1]*60 + parts[2]
        else:
            return float(t)
    except:
        return None


# -----------------------
# Extraction loop
# -----------------------

for _, row in df.iterrows():

    clue_id = int(row["clue_id"])
    round_num = row["Round"]
    date_folder = row["date_folder"]

    start = to_seconds_ms(row["Clue video start"])
    end = to_seconds_ms(row["Clue video end "])

    if start is None or end is None:
        print(f"Skipping clue {clue_id} (missing timestamp)")
        continue

    duration = end - start

    input_video = audio_root / date_folder / f"Round {round_num}.mov"

    if not input_video.exists():
        print(f"Missing video file: {input_video}")
        continue

    output_file = output_dir / f"clue_{clue_id:04d}.mp4"

    cmd = [
        "ffmpeg",
        "-y",
        "-i", str(input_video),
        "-ss", str(start),
        "-t", str(duration),
        "-c", "copy",
        str(output_file)
    ]

    subprocess.run(cmd, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

    print(f"Saved {output_file}")


# -----------------------
# Save metadata
# -----------------------

# metadata_file = "clues_metadata.csv"
# df.to_csv(metadata_file, index=False)
# 
# print(f"\nMetadata saved to {metadata_file}")
print("Done.")
```
