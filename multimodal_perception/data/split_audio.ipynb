```python
import pandas as pd
import subprocess
from pathlib import Path
```

```python
# -----------------------
# Paths
# -----------------------

csv_path = "pilot_v2.csv"                  # original csv
audio_root = Path("../../../pilot/")    # folder containing date folders
output_dir = Path("assets/audio/pilot_v2")

output_dir.mkdir(parents=True, exist_ok=True)

# -----------------------
# Load CSV
# -----------------------

df = pd.read_csv(
    csv_path,
    dtype={
        "Clue audio start": str,
        "Clue audio end ": str
    }
)

# add clue ids
df["clue_id"] = range(1, len(df) + 1)

# convert date format (27/02/2026 -> 27-02-2026)
df["date_folder"] = pd.to_datetime(df["Date "], dayfirst=True).dt.strftime("%d-%m-%Y")


# -----------------------
# Time parser
# -----------------------

def to_seconds(t):

    if pd.isna(t):
        return None

    t = str(t).strip()

    if ":" in t:
        parts = t.split(":")

        if len(parts) == 2:
            m, s = parts
            return int(m) * 60 + int(s)

        elif len(parts) == 3:
            h, m, s = parts
            return int(h) * 3600 + int(m) * 60 + int(s)

    try:
        return int(float(t))
    except:
        return None


# -----------------------
# Extraction loop
# -----------------------

for _, row in df.iterrows():

    clue_id = int(row["clue_id"])
    round_num = row["Round"]
    date_folder = row["date_folder"]

    start = to_seconds(row["Clue audio start"])
    end = to_seconds(row["Clue audio end "])

    if start is None or end is None:
        print(f"Skipping clue {clue_id} (missing timestamp)")
        continue

    duration = end - start

    input_audio = audio_root / date_folder / f"Round {round_num}.m4a"

    if not input_audio.exists():
        print(f"Missing audio file: {input_audio}")
        continue

    output_file = output_dir / f"clue_{clue_id:04d}.m4a"

    cmd = [
        "ffmpeg",
        "-y",
        "-i", str(input_audio),
        "-ss", str(start),
        "-t", str(duration),
        "-c", "copy",
        str(output_file)
    ]

    subprocess.run(cmd, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

    print(f"Saved {output_file}")


# -----------------------
# Save metadata
# -----------------------

# metadata_file = "clues_metadata.csv"
# df.to_csv(metadata_file, index=False)
# 
# print(f"\nMetadata saved to {metadata_file}")
print("Done.")
```
